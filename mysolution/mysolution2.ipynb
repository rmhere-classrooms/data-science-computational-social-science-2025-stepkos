```python
import igraph as ig
import matplotlib.pyplot as plt
```

# 2. Generowanie grafu Barabasi-Albert z 1000 węzłów

```python
n = 1000  # liczba węzłów
m = 3  # liczba krawędzi dodawanych dla nowego węzła
g = ig.Graph.Barabasi(n=n, m=m, directed=False)
```

# 3. Wizualizacja grafu layoutem Fruchterman-Reingold

```python
layout = g.layout("fr")
fig, ax = plt.subplots(figsize=(10, 10))
ig.plot(
    g,
    target=ax,
    layout=layout,
    vertex_size=5,
    vertex_label=None,
    edge_color="gray"
)
plt.show()
```

# 4. Znalezienie najbardziej centralnego węzła wg betweenness

```python
betweenness = g.betweenness()
most_central_node = betweenness.index(max(betweenness))
print(f"Najbardziej centralny węzeł (betweenness): {most_central_node}")
print(f"Wartość betweenness: {betweenness[most_central_node]:.4f}")
```

# 5. Obliczenie średnicy grafu

```python
diameter = g.diameter()
print(f"Średnica grafu: {diameter}")
```

# 6. Komentarz:
Grafy Barabasi-Albert i Erdos-Renyi różnią się istotnie w sposobie tworzenia:

## Barabasi-Albert:
tworzy grafy skalowane zgodnie z prawem potęgowym (power-law),
czyli kilka węzłów ma dużo połączeń (huby), a większość ma mało połączeń. Powstaje sieć z "hubami".

## Erdos-Renyi:
tworzy graf losowy, każda para węzłów jest połączona z jednakowym prawdopodobieństwem, rozkład stopni węzłów jest bliski rozkładowi Poissona, brak wyraźnych hubów.
